```python
import numpy as np
import pylab

import lalsimulation as lalsim
import pycbc
from pycbc import waveform
```

```python
# In PyCBC there are functions to compute directly the match between different waveforms

# First let's create two waveforms

hp1, hc1 = waveform.get_td_waveform(approximant="IMRPhenomPv2",
                                 mass1=19, mass2=11,
                                  spin1z=0.5, spin2z=-0.5,
                                  delta_t=1.0/4096,
                                  f_lower=20)

hp2, hc2 = waveform.get_td_waveform(approximant="IMRPhenomPv2",
                                 mass1=24, mass2=13,
                                  spin1z=0.2, spin2z=-0.2,
                                  delta_t=1.0/4096,
                                  f_lower=20)

pylab.plot(hp2.sample_times, hp2,label='wf 1')
pylab.plot(hp1.sample_times, hp1,label='wf 2')
pylab.xlim([-0.1,0.05])
pylab.legend()
pylab.show()
```

```python
#PSD
psd = hp1.psd(4)

# Computing the match is easy

match = hp1.match(hp2)
#match = pycbc.filter.matchedfilter.match(hp1,hp2)

print(match[0])

print("The match is low because these are very different waveforms.")
```
